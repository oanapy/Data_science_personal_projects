```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
%matplotlib inline

#https://www.datahubbs.com/machine-learning-supply-chain-forecasting-3/
```

```python
url = 'https://www.datahubbs.com/wp-content/uploads/2017/03/timeseries.csv'
```

```python
df = pd.read_csv(url)
# convert date to datetime
df['date'] = pd.to_datetime(df['date'])
df.head()
```

```python
n = len(df)
#the train and test sets are created by not randomly picking out values because a timeseries analysis is a process as a function of time
#therefore we need to allow for that process to be contained within the data selected (75% of the first rows of the dataset)
train_count = round(n * 0.75)
train = df.iloc[:train_count]
test = df.iloc[train_count + 1:n]
```

```python
#naive forecast -  assumes that the previous period's demand is going to be the next period's demand. 
#For some data, that works well enough (the best predictor of tomorrow's weather is today's weather) because changes may
#be slow. More than anything though, it serves as a quick baseline, although often unsatisfcatory in many cases.

y_hat = test.copy().drop('sales', axis=1)
y_hat['date'] = y_hat['date'] + np.timedelta64(1, "D")
y_hat['naive'] = train.loc[len(train)-1, 'sales']


plt.figure(figsize=(12,8))
plt.plot(train['date'], train['sales'], label='Train')
plt.plot(test['date'], test['sales'], label='Test')
plt.plot(y_hat['date'], y_hat['naive'], label='Naive Forecast')
plt.legend(loc='best')
plt.title("Naive Forecast")
plt.show()
```

```python
#evaluating naive forecast error
#mean absolute error 
#root mean square error
#mean absolute percentage error

evaluation = pd.merge(test, y_hat, on='date')
evaluation.columns = ['date', 'sales', 'naive']
evaluation['error'] = evaluation['sales'] - evaluation['naive']
evaluation.insert(0, 'series', 1) # insert value to groupby
evaluation.groupby('series').agg({
        'sales' : 'sum',
        'naive' : 'sum',
        'error': {
            'total_error' : 'sum',
            'mae': lambda x: np.mean(np.abs(x)),
            'rmse': lambda x: np.sqrt(np.mean(x ** 2)),
            'mape': lambda x: 100 * np.sum(np.abs(x)) / np.sum(evaluation['sales'])
        }})
```

```python
#seasonal naive
#This is a step up from our naive forecast but takes seasonal patterns into account by looking at what happened at the 
#same time last year. For example, you want to predict your sales on Christmas 2017, the seasonal naive will assume 
#same number of sales for Christmas 2017 as you had on Christmas 2016. 
#Obviously, for this to work, you must have at least one-year of data.

dates = (test['date'] - np.timedelta64(1, 'Y')).values.astype('datetime64[D]')
seasonal_naive = train[train['date'].isin(dates)]['sales'] # seasonal naive prediction
y_hat_sn = test.copy().drop('sales', axis=1)
y_hat_sn['seasonal_naive'] = pd.DataFrame(seasonal_naive).set_index(test.index)
y_hat_sn.head()


plt.figure(figsize=(12,8))
plt.plot(train['date'], train['sales'], label='Train')
plt.plot(test['date'], test['sales'], label='Test')
plt.plot(y_hat_sn['date'], y_hat_sn['seasonal_naive'], label='Seasonal Naive')
plt.legend(loc='best')
plt.show()
```

```python
#evaluation metrics for seasonal naive

evaluation = pd.merge(test, y_hat_sn, on='date')
evaluation.columns = ['date', 'sales', 'seasonal_naive']
evaluation['error'] = evaluation['sales'] - evaluation['seasonal_naive']
evaluation.insert(0, 'series', 1) # insert value to groupby
evaluation.groupby('series').agg({
        'sales' : 'sum',
        'seasonal_naive' : 'sum',
        'error': {
            'total_error' : 'sum',
            'mae': lambda x: np.mean(np.abs(x)),
            'rmse': lambda x: np.sqrt(np.mean(x ** 2)),
            'mape': lambda x: 100 * np.sum(np.abs(x)) / np.sum(evaluation['sales'])
        }})
```

```python
#average forecast - assumes the average of all of your training data
y_hat_avg = test.copy().drop('sales', axis=1)
y_hat_avg['avg_forecast'] = train['sales'].mean()

plt.figure(figsize=(12,8))
plt.plot(train['date'], train['sales'], label='Train')
plt.plot(test['date'], test['sales'], label='Test')
plt.plot(y_hat_avg['date'], y_hat_avg['avg_forecast'], label='Average Forecast')
plt.legend(loc='best')
plt.show()
```

```python
#evaluation metrics for average sales

evaluation = pd.merge(test, y_hat_avg, on='date')
evaluation.columns = ['date', 'sales', 'average_sales']
evaluation['error'] = evaluation['sales'] - evaluation['average_sales']
evaluation.insert(0, 'series', 1) # insert value to groupby
evaluation.groupby('series').agg({
        'sales' : 'sum',
        'average_sales' : 'sum',
        'error': {
            'total_error' : 'sum',
            'mae': lambda x: np.mean(np.abs(x)),
            'rmse': lambda x: np.sqrt(np.mean(x ** 2)),
            'mape': lambda x: 100 * np.sum(np.abs(x)) / np.sum(evaluation['sales'])
        }})
```

```python
#drift method - variation of the naive forecast that allows the forecast to 'drift' higher or lower based on the historical data

# Get the slope
y_t = train.loc[len(train)-1]['sales'] 
m = (y_t - train.loc[0]['sales']) / len(train)
y_hat_drift = test.copy().drop('sales', axis=1)
h = np.linspace(0,len(y_hat_drift)-1, len(y_hat_drift))
y_hat_drift['drift_forecast'] = y_t + m * h

plt.figure(figsize=(12,8))
plt.plot(train['date'], train['sales'], label='Train')
plt.plot(test['date'], test['sales'], label='Test')
plt.plot(y_hat_drift['date'], y_hat_drift['drift_forecast'], 
        label='Drift Forecast')
plt.legend(loc='best')
plt.show()
```

```python
#linear regression
from sklearn.linear_model import LinearRegression

lm = LinearRegression()
X = np.atleast_2d(train.index).T
y = np.atleast_2d(train.sales).T
lm.fit(X, y)

y_hat_reg = test.copy().drop('sales', axis=1)
y_hat_reg['lin_forecast'] = lm.predict(np.atleast_2d(y_hat_reg.index).T)

plt.figure(figsize=(12,8))
plt.plot(train['date'], train['sales'], label='Train')
plt.plot(test['date'], test['sales'], label='Test')
plt.plot(y_hat_reg['date'], y_hat_reg['lin_forecast'], label='Linear Regression')
plt.legend(loc='best')
plt.show()
```

```python
#evaluation aggregated for all timeseries methods

# Merge each of the forecasts on the date column
evaluation = pd.merge(test, y_hat, on='date')
evaluation = evaluation.merge(y_hat_sn, on='date').merge(
    y_hat_avg, on='date').merge(
    y_hat_drift, on='date').merge(
    y_hat_reg, on='date')
evaluation.insert(0, 'series', 1) # insert value to groupby

# Define metric calculations
fcast_metrics = {'total' : 'sum',
            'total_error' : lambda x: np.sum(evaluation['sales'] - x),
            'mae' : lambda x: np.mean(np.abs(x - evaluation['sales'])),
            'rmse': lambda x: np.sqrt(np.mean((x - evaluation['sales']) ** 2)),
            'mape': lambda x: 100 * np.sum(np.abs(x - evaluation['sales'])
                                          ) / np.sum(evaluation['sales'])
        }

# Create a list of the data frame objects
forecast_df = [y_hat, y_hat_sn, y_hat_avg, y_hat_drift, y_hat_reg]
# create a list of the forecast names - note that these must match the names
# used in the data frames
forecasts = ['naive', 'seasonal_naive', 'avg_forecast', 'drift_forecast',
             'lin_forecast']
# Initialize empty data frame
metric_data = pd.DataFrame()

# Loop through the forecast data frames and data to recombine and calculate
# each metric for each forecast
for f, fname in zip(forecast_df, forecasts):
    metric_calc = pd.merge(test, f, on='date')
    metric_calc.insert(0, 'series', 1)
    metric_calc = metric_calc.groupby('series').agg({fname:fcast_metrics})
    metric_calc = metric_calc.xs(fname, axis=1, drop_level=True).reset_index()
    metric_data = pd.concat([metric_data, metric_calc])

# drop the 'series' placeholder used for grouping
metric_data = metric_data.drop('series', axis=1)
# insert forecast names
metric_data.insert(0,'forecast', forecasts)
metric_data
```

```python
#timeseries breakdown as a different method to look at timeseries analysis

url = 'https://www.datahubbs.com/wp-content/uploads/2017/05/decomposition.csv'
data = pd.read_csv(url, sep=',')
# Label columns
data.columns = ['date', 'y']
# Set dates as index
data = data.set_index(pd.to_datetime(data.iloc[:,0])).drop('date', axis=1)
data.head()
```

```python
plt.figure(figsize=(12,8))
plt.plot(data['y'])
plt.show()
```

```python
# Split on Jan 1, 2017
train = data.loc[data.index < '2017-01-01']
test = data.loc[data.index > '2017-01-01']
```

```python
# Monthly cycle during the year
df_year = pd.DataFrame({"y": train.y.resample("M").sum()})
df_year['mva'] = df_year.y.rolling(center=True, window=12).mean()

plt.figure(figsize=(12,8))
plt.plot(df_year['y'], label='Actual')
plt.plot(df_year['mva'], label='Monthly Average')
#mva = monthly average
plt.legend(loc='best')
plt.show()
```

```python
#sr = seasonal ratio
df_year['sr'] = df_year['y'] / df_year['mva']

#With the SR in hand, we can now turn to calculating the Unadjusted Seasonal Index (USI). 
#If we're looking at monthly data, then this is an average of all the Septembers in our data set so that each September 
#can be adjusted individually according to the seasonality seen in September. Obviously, the same goes for all of the 
#other months in our data set as they need to be individually adjusted. The common way of doing this is by averaging each
#of the months across our time series to obtain an average by which to adjust by. 
```

```python
# Add month numbers
df_year['month'] = df_year.index.month
# Extract all January data for example
df_ = df_year.loc[df_year['month']==1]
df_
```

```python
#To get the ASI for January ( USI1 ), we need to average the seasonal ratios for January. 
#Of course, we are missing one month because of the moving average, so that will be excluded from our analysis.
#This process is then repeated for each month in the data set to provide  USIi  for each time period. 

# Average each month 
df_ = df_year.groupby('month').agg({
        'sr': 'mean'})
df_.reset_index(inplace=True)
df_.columns = ['month', 'usi']

# Combine with main data frame
df_year = pd.merge(df_year, df_, on='month', right_index=True).sort_index()

df_year['asi'] = df_['usi'].mean() * df_year['usi']

df_year['y_t-s'] = df_year['y'] / df_year['asi']
```

```python
plt.figure(figsize=(12,8))
plt.plot(df_year['y'], label='$y$')
plt.plot(df_year['y_t-s'], label='$y_{t-s}$')
plt.legend(loc='best')
plt.show()

#As you can see in the plot above, the large cycles in the sales have been smoothed out and lessened by our 
#de-seasonalizing efforts. There is an obvious upward trend that's occurring, which is what we'll focus on next.
```

```python
# trend approach
#This trend could be non-linear as well, at which point we may need to do a transformation to fit our linear model 
#Technically, we're going to be calculating the trend-cycle ( TtCt ) component here rather than strictly calculating the 
#trend itself. This can be broken up if you see an inflection point in your data where it seems like the trend has changed,
#which may indicate a structural change or a cycle.
from sklearn.linear_model import LinearRegression

lm = LinearRegression(normalize=False, fit_intercept=True)
y_t_s = np.atleast_2d(df_year['y_t-s'].values).T
x = np.atleast_2d(np.linspace(0, len(df_year) - 1, len(df_year))).T
lm.fit(x, y_t_s)
df_year['trend'] = lm.predict(x)
```

```python
# Plot actual data, de-seasonalized data, and the trend
plt.figure(figsize=(12,8))
plt.plot(df_year['y'], label='$y$')
plt.plot(df_year['y_t-s'], label='$y_{t-s}$')
plt.plot(df_year['trend'], label="$T'$")
plt.legend(loc='best')
plt.show()
```

```python
df_year['noise'] = (df_year['y'] / (df_year['asi'] * df_year['trend'])).mean()
df_year.head()
```

```python
test_year = pd.DataFrame({"y": test.y.resample("M").sum()})
test_year['month'] = test_year.index.month

# get index for trend regression
x_test = np.linspace(len(df_year), len(df_year) + 
                     len(test_year) - 1,
                    len(test_year)).reshape(-1,1)
```

```python
df_test = pd.merge(test_year, df_year[['month', 'asi', 'noise']], 
                   on='month', 
                   right_index=True).sort_index().drop_duplicates()
df_test['trend'] = lm.predict(x_test)
df_test['forecast'] = df_test['asi'] * df_test['noise'] * df_test['trend']
df_test
```

```python
plt.figure(figsize=(12,8))
plt.plot(df_year['y'], label='Train ($y_t$)')
plt.plot(df_test['y'], label='Test ($y_t$)')
plt.plot(df_test['forecast'], label='Forecast ($\hat{y_t}$)')
plt.legend(loc='best')
plt.title("Classical Decomposition and Multiplicative Model Forecast")
plt.show()
```

```python
evaluation = df_test.copy()
evaluation['error'] = evaluation['y'] - evaluation['forecast']
evaluation.insert(0, 'series', 1) # insert value to groupby
evaluation.groupby('series').agg({
        'y' : 'sum',
        'forecast' : 'sum',
        'error': {
            'total_error' : 'sum',
            'percentage_error' : lambda x: 100 * np.sum(x) / np.sum(evaluation['y']),
            'mae': lambda x: np.mean(np.abs(x)),
            'rmse': lambda x: np.sqrt(np.mean(x ** 2)),
            'mape': lambda x: 100 * np.sum(np.abs(x)) / np.sum(evaluation['y'])
        }}).apply(np.round, axis=1)


#MAPE is 6% - which is pretty good for the whole protocol of forecasting
```
